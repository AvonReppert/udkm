```python
offset = -0.2
plt.figure()
for i, line in enumerate([1,2,3,4]):
    params = moke.get_scan_parameter(parameter_file_name, line)
    params["bool_t0_shift"] = True
    params["bool_save_plot"] = True
    params["t0_column_name"] = "field_up"
    params["plot_path"] = "plot_overview//"
    params["scan_path"] = "scan_export//"
    
    scan = moke.load_overview_data(params)
    plt.plot(scan["delay"], scan["field_up"] + i*offset, label = scan["sample"],lw = 2)
plt.xlabel("delay t (ps)")
plt.ylabel("reflectivity (arb. units)")
plt.legend(loc=1,ncol = 2)
plt.xlim((-10, 300))
plt.ylim((-1,0.1))
plt.title("reflectivity")
plt.show()
```
