# plot module

the plot module provides a set of predefined plot function that can come in handy for a quick overview plot that can be further customized. In addition udkm python library also provides a matplotlib base style that can be imported to change the default matplotlib plot style with more common plot layouts for publication in two-column papers.

## usage of the matplotlib style file

In udkm.tools you find the file ```udkm_base.mplstyle``` that we use as a default in matplotlib.
To add it to the preinstalled matplotlib styles follow these steps:

1. Look for the function path of matplotlib using
```
import matplotlib
matplotlib.matplotlib_fname()
```
2. There you will find a folder called ```stylelib``` to which you copy the file ```udkm_base.mplstyle``` 
3. After restarting your kernel you should be able to use the new plotstyle via:
```
import matplolitb.pyplot as plt
plt.style.use("udkm_base")
``` 
Alternatively you can import the plotstyle file into your runtime using the absolute or relative path
to the style file via ```plt.style.use(path_to_file+"/udkm_base.mplstyle")```, potentially also in the startup routine of your python environment.

````python
## import and initialization of relevant functions
import os
import numpy as np
import matplotlib.pyplot as plt

import udkm.tools.colors as colors
import udkm.moke.functions_2 as moke
import udkm.tools.plot as plot

cwd = os.getcwd()


x = np.arange(0,10,0.1)

## example for a plot using the default matplotlib style
plt.style.use("default")

plt.figure()
plt.plot(x, np.sin(x), label="sin(x)")
plt.plot(x, np.cos(x), label="cos(x)")
plt.plot(x, np.tan(x), label="tan(x)")
plt.plot(x, 0.2*np.cosh(x), label="cosh(x)/5")
plt.axhline(y= 0, ls = "--",  color = "grey")
plt.xlabel("x"); plt.ylabel("f(x)")
plt.legend(loc = 1)
plt.ylim(-5,5)
plt.xlim()
plt.title("python default plot")
plt.show()

## example for a plot using the udkm matplotlib style

plt.style.use("../../../udkm/tools/udkm_base.mplstyle")

plt.figure()
plt.plot(x, np.sin(x), label="sin(x)")
plt.plot(x, np.cos(x), label="cos(x)")
plt.plot(x, np.tan(x), label="tan(x)")
plt.plot(x, 0.2*np.cosh(x), label="cosh(x)/5")
plt.xlabel("x"); plt.ylabel("f(x)")
plt.axhline(y= 0, ls = "--", color = "grey")
plt.legend()
plt.title("udkm matplotlib style")
plt.ylim(-5,5)
plt.xlim(0,5)
plt.show()

````

## pre-defined plot functions


````python
# loading sample data
example_data_path='../../../udkm/moke/example_2/'
os.chdir(example_data_path)

sample_name = 'P28b'
reference_file = 'fluence_series_2.txt'

analysis_params = moke.set_analysis_params(sample_name)
analysis_params["bool_plot_overview"] = False


series = moke.load_series(reference_file, analysis_params)
````

### list_plot
provides a quick overview of a measurement series and is suited for qualitative and quantitative comparison when the data do not overlap

````python
plt.figure()
ax1 = plot.list_plot(series['delay'], series['moke'], np.round(series['fluence'],1),cmap = colors.cmap_blue_red_2)

plt.xlabel("delay (ps)")
plt.ylabel("moke (V)")
plt.axhline(y=0, ls="--", color=colors.grey_2)
plt.legend(fontsize=8, loc=4,
           title=r"Fluence F $\left(\,\mathrm{\frac{mJ}{\,cm^2}}\right)$", handlelength=1, ncol=3)
plt.title("list_plot example")
plt.show()
````

### offset_plot
provides qualitative overview of a measurement series. It is useful when the data overlap so that a list plot would be too crowded

````python
plt.figure()
ax2 = plot.offset_plot(series['delay'], series['moke'], np.round(series['fluence'],1),
                 offset = -0.4,cmap = colors.cmap_red_blue_2, x_text = 80, y_text = 0.05)

ax2.set_xlabel("delay (ps)")
ax2.set_ylabel("moke (V)")
ax2.axhline(y=0, ls="--", color=colors.grey_2)
ax2.set_ylim((-5.5,0.75))
plt.title("offset_plot example")
ax2.annotate(r"F $ \mathrm{ \left( \frac{mJ}{\,cm^2} \right)} $ ", xy=(100, 0.33), xycoords='data',
                     xytext=(100, 0.33), fontsize=10, color=colors.grey_1)
plt.show()
````
